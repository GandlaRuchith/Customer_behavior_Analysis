```python

from sqlalchemy import create_engine, text

# --- SQL Server connection details ---
server = '(localdb)\MSSQLLocalDB'             # or 'localhost\SQLEXPRESS'
database = 'DataAnalysis'
driver = 'ODBC Driver 18 for SQL Server'  # replace with your installed driver

# Create SQLAlchemy engine with Windows Authentication
connection_string = f'mssql+pyodbc://@{server}/{database}?driver={driver.replace(" ", "+")}&trusted_connection=yes'
engine = create_engine(connection_string)

# --- Test connection ---
try:
    with engine.connect() as conn:
        result = conn.execute(text("SELECT 1"))
        print("Connection successful:", result.fetchone())
except Exception as e:
    print("Connection failed:", e)

# --- Upload your DataFrame to SQL Server ---
try:
    df.to_sql(
    name='customer_shopping_behavior',
    con=engine,
    schema='dbo',
    if_exists='append',
    index=False
)
    print("DataFrame upload successful")
except Exception as e:
    print("Upload failed:", e)
```
